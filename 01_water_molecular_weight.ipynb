```python
from rdkit import Chem
from rdkit.Chem import Draw

# A shorthand text representation for Aspirin
aspirin_smiles = "CC(=O)OC1=CC=CC=C1C(=O)O"

# Convert that text line into an interactive chemical object
aspirin_molecule = Chem.MolFromSmiles(aspirin_smiles)

# Render the molecule dynamically onto your screen
Draw.MolToImage(aspirin_molecule)
```
