```python
raw_bioactivity_data = [
    {
        'molecule_name': 'Gefitinib',
        'canonical_smiles':'COCCOCCOCCO...',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 35, 
            'unit': 'nM'
        }
    },
    {
        'molecule_name': 'Aspirin',
        'canonical_smiles': 'CC(=O)OC1=CC=CC=C1C(=O)O)',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 5000,
            'unit': 'nM'
        }
    },
    {
        'molecule_name': 'Osimertinib',
        'canonical_smiles': 'CN(C)CC=CC(=O)N...',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 15,
            'unit': 'nM'
        }
    }
]
for compound in raw_bioactivity_data :
    details = compound.get('bioactivity_details', {})
    bio_type = details.get('type', 'Unknown')
    bio_value = details.get('value', 999)
    if bio_type == "IC50" and bio_value < 100:
        name = compound.get('molecule_name', 'Nameless')
        print(f"DETECT QUALIFIED COMPOUND: {name:<20} | IC50_value = {bio_value} ")
```

```python
messy_bioactivity_data = [
    {
        'molecule_name': 'Gefitinib',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 35, 
            'unit': 'nM' # Already in nM!
        }
    },
    {
        'molecule_name': 'Aspirin',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 5.2,
            'unit': 'uM' # <-- Needs to be converted! (5.2 * 1000 = 5200 nM)
        }
    },
    {
        'molecule_name': 'Osimertinib',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 0.015,
            'unit': 'uM' # <-- Needs to be converted! (0.015 * 1000 = 15 nM)
        }
    }
]
for compound in messy_bioactivity_data:
    details = compound.get('bioactivity_details', {})
    bio_value = details.get('value', 999)
    bio_type = details.get('unit', {})
    name = compound.get('molecule_name', 'nameless')
    if bio_type == "uM":
        bio_value = bio_value * 1000
        bio_type = "nM"
    if bio_value < 100 :
        print(f"FINAL COMPOUND {name:<20} | IC50 value: {bio_value:.2f}")
        
```

```python
from chembl_webresource_client.new_client import new_client

try:
    # 1. Access target API
    target_api = new_client.target
    
    # 2. Search for human EGFR
    print("Connecting to ChEMBL in Europe...")
    results = target_api.filter(target_organism="Homo sapiens", target_type="SINGLE PROTEIN").search("EGFR")
    
    # 3. Try to pull the first match
    first_match = list(results)[0]
    print("\n✅ CONNECTION RECOVERED!")
    print(f"Target Name: {first_match.get('pref_name')}")
    print(f"ChEMBL ID:   {first_match.get('target_chembl_id')}")

except Exception as error:
    print("\n❌ Server is still undergoing maintenance.")
    print(f"Error details: {error}")
```
