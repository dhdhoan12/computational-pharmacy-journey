```python
import pandas as pd
messy_compounds = [
    {
        'molecule_name': 'Gefitinib',
        'bioactivity_details': { 
            'type': 'IC50',
            'value': 35,
            'unit': 'nM'
        }
    },
    {
        'molecule_name': 'Asipirin',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 5.2,
            'unit': 'uM'
        }
    },
    {
        'molecule_name': 'Osimertinib',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 0.015,
            'unit': 'uM'
        }
    },
    {
        'molecule_name': 'Lapatinib',
        'bioactivity_details': {
            'type': 'IC50',
            'value': 120,
            'unit': 'nM'
        }
    }
]
cleaned_list = []
for compound in messy_compounds:
    details = compound.get('bioactivity_details', {})
    bio_value = details.get('value', 999)
    unit = details.get('unit', 'Unknown')
    name = compound.get('molecule_name', 'Nameless')
# Standarlize uM to nM
    if unit == 'uM':
        bio_value = bio_value * 1000
        unit = 'nM'
    if bio_value < 100:
        cleaned_record = {
            'Molecule_name': name,
            'IC50 (nM)': round(bio_value, 2),
            'Status': 'Potent'
        }
        cleaned_list.append(cleaned_record)
df = pd.DataFrame(cleaned_list)
df
    

            
```

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client

print("Step 1: Connecting to ChEMBL API Server")
activities = new_client.activity
raw_query = activities.filter(target_chembl_id = "CHEMBL203").filter(standard_type = "IC50")

print("Step 2: Downloading an expansive slice of 1000 raw records")
egfr_large_list = list(raw_query[:1000])
df_raw = pd.DataFrame(egfr_large_list)

print("Step 3: cleaning noise and missing values...")
df_subset = df_raw[['molecule_chembl_id', 'standard_value', 'standard_units']].copy()
df_clean = df_subset.dropna(subset=['standard_value']).copy()
df_clean['standard_value'] = pd.to_numeric(df_clean['standard_value'])

print("Step 4: Sifting for potent drug inhibitors (<100nM)... ")
potent_compounds = df_clean[df_clean['standard_value'] < 100]
# Deduplicate
unique_count = potent_compounds['molecule_chembl_id'].nunique()
print("Done")
print(f"Total raw records evaluated: {len(df_raw)}")
print(f"Potent entries remaining: {len(potent_compounds)}")
print(f"Total unique molecules discovered: {unique_count}")

potent_compounds.to_csv("egfr_potent_inhibitors.csv", index = False)
print("File successfully saved as: 'egfr_potent_inhibitors.csv' ")
potent_compounds.head()

```
